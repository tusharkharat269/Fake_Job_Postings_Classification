## What are the most common title used in jobs in the US?

## Which department has the most number of fake jobs?

## Which department or function has high-paying jobs in the UK?


```python
import matplotlib.pyplot as plt

# Assuming 'salary_data' is your DataFrame containing job titles, salary ranges, and locations

# Filter data for UK-related locations
uk_data = df[df['location'].isin(['GB', 'LND', 'London'])]

# Define a function to convert salary ranges to a numerical representation (midpoint)
def convert_salary_range(salary_range):
    if pd.notnull(salary_range):
        low, high = map(float, salary_range.split('-'))
        return (low + high) / 2
    else:
        return None

# Apply the conversion function to create an 'Average Salary' column
uk_data.loc[:, 'Average Salary'] = uk_data['salary_range'].apply(convert_salary_range)


# Group data by department and calculate the average salary
average_salary_by_department = uk_data.groupby('department')['Average Salary'].mean()

# Sort departments by average salary in descending order
sorted_departments = average_salary_by_department.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 6))
sorted_departments.plot(kind='bar', color='skyblue')
plt.xlabel('Department')
plt.ylabel('Average Salary')
plt.title('Average Salary by Department in the UK')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility

# Show the plot
plt.show()
```

## What are the top 3 most commonly used words in Company Profile?

```python
import pandas as pd
from collections import Counter
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Assuming 'df' is your DataFrame containing the 'Company Profile' column
# If you haven't already, you might need to install the NLTK library and download the stopwords data.

# Download stopwords (if you haven't done so before)
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# Get the stopwords
stop_words = set(stopwords.words('english'))

# Tokenize and clean text
def tokenize_and_clean(text):
    if pd.notnull(text):  # Check if the value is not NaN
        words = word_tokenize(text.lower())
        return [word for word in words if word.isalnum() and word not in stop_words]
    else:
        return []


# Apply tokenization and cleaning to 'Company Profile' column
df['company_profile'] = df['company_profile'].apply(tokenize_and_clean)

# Flatten the list of words
all_words = [word for words_list in df['company_profile'] for word in words_list]

# Count the frequency of each word
word_counts = Counter(all_words)

# Get the top 3 words
top_3_words = word_counts.most_common(3)

print(top_3_words)
```

```python
from bs4 import BeautifulSoup
import spacy

stop_words = spacy.lang.en.stop_words.STOP_WORDS

#function to lemmatize the text
load_model = spacy.load('en_core_web_sm',disable = ['parser','ner'])
def lemmatize(text):
    doc = load_model(text)
    tex= " ".join([token.lemma_ for token in doc])
    return tex         


#function to clean the whole text
def clean_text(text):
    #Converting to lowercase
    text = text.lower()
    #Removing emails
    text = re.sub(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9._-]+\.[a-zA-Z0-9_-]+)', '', text)
    #Removing URLs
    text = re.sub(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', text)
    #Removing HTML tags
    text = BeautifulSoup(text, 'lxml').get_text()
    #Removing punctuations and numbers
    text = re.sub('[^A-Z a-z ]+', ' ', text)
    #Removing Multiple spaces
    text =  " ".join(text.split())
    #Removing Stop words
    text =  " ".join([t for t in text.split() if t not in stop_words])
    #lemmatizing the text
    text = lemmatize(text)
    return text    
```

# spitting Train and Test Data

```python

# Initialize a TfidfVectorizer with a maximum of 100 features
cv = TfidfVectorizer(max_features=100)

# Transform the 'text' column in the DataFrame 'df' into a TF-IDF matrix
x = cv.fit_transform(df['text'])

# Create a DataFrame 'ds1' from the TF-IDF matrix with feature names
ds1 = pd.DataFrame(x.toarray(), columns=cv.get_feature_names_out())

# Combine the TF-IDF DataFrame 'ds1' with the 'fraudulent' column from the original DataFrame 'df'
clean_df = pd.concat([ds1, df.fraudulent], axis=1)

# Display the first few rows of the combined DataFrame 'clean_df'
clean_df.head()
```

```python
y = clean_df.iloc[:,-1]
X = clean_df.iloc[:,:-1]

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Creating model

### SVM

```python
from sklearn.svm import SVC

svm = SVC(C=1 ,kernel = 'linear', random_state = 1)
model_svm = svm.fit(X_train,y_train)
```

```python
pred_svm = svm.predict(X_test)

accuracy_svm = accuracy_score(y_test,pred_svm) 
print(accuracy_svm)
```

```python
print("Report")
print(classification_report(y_test,pred_svm))
print("Confusion Matrix\n")
print(confusion_matrix(y_test,pred_svm))
```

### Summary:
- The model performs very well at predicting class 0, achieving high precision, recall, and F1-score.

- However, the model struggles to predict class 1. It has low precision, recall, and F1-score for this class, suggesting that it may need further optimization or that the class is imbalanced.

- It's important to note that class 1 has very few instances (only 254) compared to class 0 (5110), which could contribute to the poor performance for class 1. Depending on the context, you might consider strategies like resampling techniques or using different models to improve performance for the minority class.

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_jobs = 3,oob_score = True,n_estimators=100,criterion = 'entropy')
model_rfc = rfc.fit(X_train,y_train)
```

```python
pred_rfc = rfc.predict(X_test)

accuracy_rfc = accuracy_score(y_test,pred_rfc) 
print(accuracy_rfc)
```

```python
print("Report")
print(classification_report(y_test,pred_rfc))
print("Confusion Matrix\n")
print(confusion_matrix(y_test,pred_rfc))
```

## Ensemble model

```python
from sklearn.ensemble import VotingClassifier

# Create a list of tuples with model names and the corresponding models
models = [('SVM', svm), ('KNN', knc), ('Random Forest', rfc)]

# Initialize Voting Classifier
ensemble_model = VotingClassifier(estimators=models, voting='hard')  # 'hard' for probability voting

# Train the ensemble model
ensemble_model.fit(X_train, y_train)

# Make predictions
y_pred = ensemble_model.predict(X_test)

# Evaluate the model
from sklearn.metrics import classification_report

report = classification_report(y_test, y_pred)
print(report)
```
